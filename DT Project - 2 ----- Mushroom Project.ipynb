```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in df.columns:
    df[i] = le.fit_transform(df[i])
    
df
```

As all the columns were in text, hence did encoding using LabelEncoder for each text column to convert it into number

```python
x = df.drop('class',axis=1)
y = df['class']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split( x, y, test_size=0.25, random_state=25 )
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean( pred_i != y_test ))

plt.figure( figsize = (10,6) )
plt.plot( range(1,40), error_rate )
```

Fetching error rate for k = 1 to 40 and will check K value where error rate is minimum.
Seems like error rate is minimu i.e. 0 at K = 5

```python
error_rate
```

```python
error_rate = []
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
error_rate.append(np.mean( y_pred != y_test ))

error_rate
```

Just confirming the error rate by entering K = 5

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_pred))
print('Test accuracy:-',accuracy_score(y_test,y_pred))
```

Accuracy is 100 %, also no False Positive or False Negative, seems Model is working fine.

```python
import numpy as np
df_pred = np.array(y_pred)
df_actual = np.array(y_test)
analysis = pd.DataFrame({"Predicted Value" : df_pred, "Actual value" : df_actual, "Difference": df_pred - df_actual})
analysis
```

Comparing predicted value and actual test value by creating a dataframe.

```python
analysis["Difference"].value_counts()
```

Checking different values of Difference column, seems all values are zero, which means every preidcted value has been predicted correctly and is matching with the test value. Great !
